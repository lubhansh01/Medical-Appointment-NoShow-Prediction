```python
import pandas as pd

df = pd.read_csv("../data/medical_appointments.csv")
print(df.shape)
```

```python
# Numeric
df["age"] = df["age"].fillna(df["age"].median())

# Categorical
cat_cols = ["gender", "specialty", "disability", "place", "appointment_shift"]
for col in cat_cols:
    if col in df.columns:
        df[col] = df[col].fillna("Unknown")

# Target must NEVER be dropped
df = df[df["no_show"].notna()]
```

```python
df["no_show_binary"] = (
    df["no_show"]
    .astype(str)
    .str.strip()
    .str.lower()
    .map({"yes": 1, "no": 0})
)
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
for col in cat_cols:
    if col in df.columns:
        df[col] = le.fit_transform(df[col])
```

```python
df["appointment_date_continuous"] = pd.to_datetime(
    df["appointment_date_continuous"], errors="coerce"
)

df["day"] = df["appointment_date_continuous"].dt.day
df["month"] = df["appointment_date_continuous"].dt.month
df["weekday"] = df["appointment_date_continuous"].dt.weekday
```

```python
print("Final shape before saving:", df.shape)
print(df["no_show_binary"].value_counts())
```

```python
df.to_csv("../data/processed/classification_data.csv", index=False)
```
